Modules, Libraries and Packages

A module = a single Python file (.py) that has functions, variables, or classes.

A library = a collection of modules that helps you do tasks easily.

A package = a folder containing multiple modules (and maybe sub-packages)

Errors and Exception Handling

```python
try:
    printf(Hello)
except:
    print('Syntax issue')
```

```python
try:
    f = open('testfile.txt', 'w')
    f.write('Test write this')
except IOError:
    print("Error: couldn't find the file or read data.")
else:
    print("Content successfully written")
    f.close()
```
