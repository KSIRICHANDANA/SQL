```python
n = int(input("Enter the number of rows: "))
mid = n // 2 + 1   # middle row = 5

for i in range(1, n + 1):
    for j in range(1, n + 1):

        # middle row (Part C)
        if i == mid:
            print("*", end="")

        # top-left (Part A)
        elif i < mid and j <= i:
            print("*", end="")

        # top-right (Part B)
        elif i < mid and j >= n - i + 1:
            print("*", end="")

        # bottom-left (Part D)
        elif i > mid and j <= n - i + 1:
            print("*", end="")

        # bottom-right (Part E)
        elif i > mid and j >= i:
            print("*", end="")

        else:
            print(" ", end="")

    print()
```

```python
n = int(input("Enter the number of rows: "))

# Top half
for i in range(1, n + 1):
    print(" " * (n - i), end="")

    if i == 1:
        print("*")
    else:
        print("*" + " " * (2 * i - 3) + "*")

# Bottom half
for i in range(n - 1, 0, -1):
    print(" " * (n - i), end="")

    if i == 1:
        print("*")
    else:
        print("*" + " " * (2 * i - 3) + "*")
```

```python
n = int(input("Enter the number of rows: "))

for i in range(1, n + 1):

    # LEFT: 1 to i
    for j in range(1, i + 1):
        print(j, end="")

    # SPACES
    for s in range(2 * (n - i)):
        print(" ", end="")

    # RIGHT: i to 1
    for j in range(i, 0, -1):
        print(j, end="")

    print()
```
