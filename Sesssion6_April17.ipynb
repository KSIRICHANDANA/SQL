OOPS and File I/O

OOP (Object-Oriented Programming) is a way of writing code where we use: objects to represent real-world things

A class is a blueprint / template for the objects we are going to create

An object is a real instance of a class


Class = design of a house/ blueprint
Object = actual house built from it

A constructor is a special method in a class that runs automatically when you create an object.

Constructor = function that sets up the object when it is born

Think:

Class = blueprint
Object = house built
Constructor = “setup work when house is built”


self is the current object (the one you are working with)
self = “this exact house”

```python
class Person:
    def __init__(self,name,surname,yob):
        self.name = name
        self.surname = surname
        self.yob = yob
```

```python
alec = Person("Alec", "Baldwin",1958)
print(alec)
print("%s %s was born in %d." %(alec.name, alec.surname, alec.yob))
```

```python
kimi = Person("Kimi", "Antonelli", 2006)
print(kimi)
print("%s %s was born in %d." %(kimi.name, kimi.surname, kimi.yob))
```

Methods

Methods = functions inside a class

Function → outside class
Method → inside class (belongs to object)

```python
class Person:
    def __init__(a,name,surname,yob):
        a.name = name
        a.surname = surname
        a.yob = yob
    
    def age(a, current_year):
        return current_year - a.yob
    
    def summ(a, current_year):
        return current_year + a.yob
    
    def __str__(a):
        return "%s %s was born in %d." %(a.name, a.surname, a.yob)
```

```python
mv = Person("Max", "Verstappen", 1997)
print(mv)
print(mv.age(2026))
print(mv.summ(2026))
```

__str__ (User-friendly)

Purpose: Give a nice, readable output for users

__repr__ (Developer-friendly)

Purpose: Give a detailed, unambiguous representation
Used for: debugging, logs, development

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def __repr__(self):
        return f"Person(name='{self.name}', age={self.age})"
```

```python
p = Person("John", 25)
```

```python
print(p)
```

```python
p
```

```python
people = [Person("John", 25), Person("Emma", 30)]
print(people)
```

Inheritance


Inheritance = one class (child) getting properties and methods from another class (parent).


Think:

Parent: “Human”
Child: “Student”

A student:

can walk (from human)
can eat (from human)
AND also has study() (new)

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Bark")
```

```python
d = Dog()

d.sound()  # inherited from Animal
d.bark()   # own method
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} is starting")

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)
        self.doors = doors

    def show(self):
        print(f"Car has {self.doors} doors")
```

```python
c = Car("Toyota", 4)
c.start()
c.show()
```

Method Overriding


Method overriding happens when a child class changes a method that it inherited from the parent class.

Simple meaning: Same method name in parent and child, but child gives a different implementation.

Modify the functionality of parent method but only in child class it stays same in parent class

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")
        
class Dog(Animal):
    def sound(self):
        print("Dog barks")
        

a = Animal()
d = Dog()

a.sound()
d.sound()
```
